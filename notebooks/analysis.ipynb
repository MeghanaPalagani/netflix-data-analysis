```python
import pandas as pd

df = pd.read_csv('../data/netflix_titles.csv')

df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['director'] = df['director'].fillna('Unknown')

df['country'] = df['country'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df['type'].value_counts()
```

```python
import matplotlib.pyplot as plt

df['type'].value_counts().plot(kind='bar')

plt.title('Movies vs TV Shows on Netflix')

plt.xlabel('Type')

plt.ylabel('Count')

plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

top_countries
```

```python
top_countries.plot(kind='bar')

plt.title('Top 10 Countries Producing Netflix Content')

plt.xlabel('Country')

plt.ylabel('Count')

plt.savefig('../images/top_countries.png')

plt.show()
```

```python
yearly_content = df['release_year'].value_counts().sort_index()

yearly_content
```

```python
yearly_content.plot(kind='line', figsize=(12,5))

plt.title('Netflix Content Release Trend Over Years')

plt.xlabel('Release Year')

plt.ylabel('Number of Shows')

plt.savefig('../images/release_trend.png')

plt.show()
```

```python
df['rating'].value_counts().head(10)
```

```python
df['rating'].value_counts().head(10).plot(kind='bar')

plt.title('Most Common Netflix Ratings')

plt.xlabel('Rating')

plt.ylabel('Count')

plt.savefig('../images/ratings_chart.png')

plt.show()
```

# Final Insights

- Movies dominate Netflix content compared to TV shows.
- The United States contributes the highest amount of Netflix content.
- Netflix content production increased rapidly after 2015.
- TV-MA and TV-14 are the most common audience ratings.
